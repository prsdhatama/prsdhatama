```python
#so to search what is the 'Or' thing we need to look at the column with 2 character from order date first
all_monthly_data['Month'] = all_monthly_data['Order Date'].str[0:2]
all_monthly_data.head()
```

```python
#it filter and shows all data which in column Month give 'Or' value then make it dissapear, therefore using '!='
#there is 355 value of 'Order Date' in Order Date column, and all the row filled with 'Order Date' is actually an ampty row
all_monthly_data[all_monthly_data['Month'].str[0:2] == 'Or'].shape[0]
all_monthly_data = all_monthly_data[all_monthly_data['Month'].str[0:2] != 'Or']

all_monthly_data.head()
```

## Adding a Sales Column


```python
#After we make a 'month' column, we need to have 'sales' column, which contain [quantitiy x price = Sales] value

#We need to change the quantity ordered and price each column to integer or float to calculate it

all_monthly_data['Quantity Ordered'] = all_monthly_data['Quantity Ordered'].astype(int)
all_monthly_data['Price Each'] = all_monthly_data['Price Each'].astype(float)

all_monthly_data.head()
```

```python
all_monthly_data['Sales'] = all_monthly_data['Quantity Ordered'] * all_monthly_data['Price Each']

all_monthly_data.head()
```

```python
#to search for the best month for Sales we can use a groupby function to groupping the Sales by the month
#using sum function on groupby function, we can get a total value of sales for each month

all_monthly_data.groupby('Month')['Sales'].sum().sort_values(ascending=True)

#From the result we get that the result is on December
```
